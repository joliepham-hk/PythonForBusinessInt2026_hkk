````python
### Task 1: A list of dictionaries (1p)

You are given a list of raw transaction records, where each record is a dictionary:

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

Write a script that performs the following steps:

a) Filter out invalid transactions where the price is zero or the quantity is negative.

b) For each remaining transaction, compute a new field `total` = `price * quantity`.

c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

d) Print the region with the highest total revenue and its value using an f-string.
````

````python
# a) Filter out invalid transactions where the price is zero or the quantity is negative.

transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# create an empty list for valid transactions
valid_transactions = []

# go through each transaction
for transaction in transactions:
    
    # check if transaction is valid
    if transaction["price"] !=0 and transaction["quantity"] >= 0:
        
        # add valid transaction to the list
        valid_transactions.append(transaction)

print (valid_transactions)
````

````python
# b) for each remaining transaction, compute a new field `total` = `price * quantity`.

# loop through the valid transactions
for transaction in valid_transactions:

    # calculate 
    transaction["total"] = transaction["price"] * transaction["quantity"]

for transaction in valid_transactions:
    print(transaction)
````

````python
# c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

# create a new dictionary for revenue per region
revenue_per_region = {}

# loop through valid transactions and group them by region
for transaction in valid_transactions:

    # explanation: If this region has never been seen before
        #register it on the scoreboard with 0 first so Python doesn't crash with a KeyError 
    if transaction["region"] not in revenue_per_region:
        revenue_per_region[transaction["region"]] = 0
    
    # '+=' add to existing value --> revenue_per_region[transaction["region"]] = revenue_per_region[transaction["region"]] + transaction["total"]
    revenue_per_region[transaction["region"]] += transaction["total"]


# d) Print the region with the highest total revenue and its value using an f-string.

print (revenue_per_region)

# find the REGION with the highest revenue
highest_region = max(revenue_per_region, key=revenue_per_region.get)
    # max(revenue_per_region, : find the maximum through the dictionary revenue_per_region
    # key=revenue_per_region.get : compare by VALUE, not region NAME

# find the highest REVENUE
highest_revenue = revenue_per_region[highest_region]

# print the region with the highest total revenue and its value using an f-string
print(f"The highest revenue region is {highest_region} with {highest_revenue}")
    # print(f"...") : print value using an f-string
````

### Task 2: Creating a function (1p)

You are given the following list of daily sales figures (in euros) for a two-week period:

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

Write a function called `summarise_sales` that takes a list of daily sales values and returns a dictionary containing:

`"total"`: the sum of all daily sales

`"average"`: the mean daily sales value, rounded to two decimals

`"best_day"`: the index (starting from 1) of the day with the highest sales

`"worst_day"`: the index (starting from 1) of the day with the lowest sales

`"above_average_count"`: the number of days where sales exceeded the average

Call the function with the list above and print the returned dictionary.

````python
# write a function

def summarise_sales(sales):
    """list of daily sales"""
    
    # calculate total daily sales, store in variable 'total'
    total = sum(sales)

    #calculate average daily sales, total divided by number of days, round the answer to '2' decimal places
    average = round(sum(sales) / len(sales), 2)

    # find the best day with the highest sales
        # .index() = find which position that number is at 
        # max(sales) = find the biggest number in the list
        # add 1 because days start from 1, not 0
    best_day = sales.index(max(sales)) + 1

    # find the worst day which has the lowest sales
    worst_day = sales.index(min(sales))+1

    # find above average: how many days had sales higher than the average
    above_average_count = 0
    for daily_amount in sales:
        if daily_amount > average:
            # for every daily sales that is above average, add 1 to the counter
            above_average_count += 1

# Return all results as a dictionary

    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }
    
````

````python
# Call the function and print the returned dictionary

# Call the function with the sales data
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

result = summarise_sales(sales)

# Print the entire returned dictionary
print(result)
````

### Task 3: Loops (1p)

You are given the following list of student records:

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
```

Using `for` loops and `if` statements, perform the following:

a) Create a new list called `passing` that contains only the names of students with a grade of 50 or above. Print the list.

b) Loop through the students and print a status message for each. If the student's grade is 80 or above, print `"[name]: Excellent"`; if between 50 and 79, print `"[name]: Pass"`; otherwise print `"[name]: Fail"`.

c) Count how many students have an attendance rate below 0.80. Print the count.

d) Find the student with the highest grade using a loop (do not use `max()`). Print the student's name and grade.

e) Now find the student with the highest grade using `max()`. Print the student's name and grade.

````python
# a) Create a new list called `passing` that contains only the names of students with a grade of 50 or above. Print the list.

students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]

# create an empty list that contains only the names of students with a grade of 50 or above
passing = []

# loop through each student in the list
for student in students:

    # check if grade is 50 or above
    if student["grade"] >= 50 or student["grade"] > 50:

        # add student names to passing list
        passing.append(student)

# print 
print("Passing list:", passing)

# or print with f-string
print(f"Passing list: {passing}")


# b) Loop through the students and print a status message for each. 
# If the student's grade is 80 or above, print `"[name]: Excellent"`; 
# if between 50 and 79, print `"[name]: Pass"`; 
# otherwise print `"[name]: Fail"`.

# loop through each student in the list
for student in students:

    # check the grades
    if student["grade"] >= 80:
        
        # print
        print(f"{student["name"]}, : Excellent")

    elif student["grade"] >= 50:

        #print 
        print(f"{student["name"]}: Pass")

    else:

        #print
        print(f"{student["name"]}: Fail")

    
# c) Count how many students have an attendance rate below 0.80. Print the count.

# create a counter
count = 0 

# loop through each student
for student in students:

    # check if attendance rate is below 0.80
    if student["attendance"] < 0.80:

        # add to the counter
        count += 1

# print
print(f"Number of students with attendance rate below 0.80: {count}")


# d) Find the student with the highest grade using a loop (do not use `max()`). Print the student's name and grade.

# assume the first student has the highest grade
highest_grade_student = students[0]

# loop through the rest of the students
for student in students:

    # for each student, check if their grade is higher than highest_grade_student["grade"]
    if student["grade"] > highest_grade_student["grade"]:

        # then update highest_grade_student 
        highest_grade_student = student

# print
print(f"Student with the highest rate: {highest_grade_student["name"], highest_grade_student["grade"]}")



# e) Now find the student with the highest grade using `max()`. Print the student's name and grade.

highest_grade_student = max(students, key=lambda student: student["grade"])
# lambda: quick function without using "def" → lambda param: return_value (use with max/min/sorted)

# print
print (f"Student with the highest rate (using max): {highest_grade_student["name"]}, {highest_grade_student["grade"]}")
````

### Task 4: Class and methods (1p)

Design a class called `Portfolio` that represents an investment portfolio. The constructor is provided for you:

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
```

Each key in `holdings` is a ticker symbol (string) and each value is a dictionary with keys `"shares"` and `"price_per_share"`. Add the following methods to the class:

a) A method `add_asset(ticker, shares, price_per_share)` that adds a new asset or updates the number of shares if the ticker already exists (in which case the price should also be updated).

b) A method `total_value()` that returns the total market value of the portfolio.

c) A method `most_valuable()` that returns the ticker of the asset with the highest total value (shares × price).

d) A `__str__` method that returns a readable summary of the portfolio, including the owner's name, the number of assets, and the total value.

Create an instance, add at least three assets, and demonstrate all methods with printed output.

````python
# Design a class called Portfolio

class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}
    
# a) A method add_asset(ticker, shares, price_per_share)
    def add_asset(self, ticker, shares, price_per_share):
        
        # adds a new asset or updates if ticker already exists
        self.holdings[ticker] = {
            "shares": shares,
            "price_per_share": price_per_share
        }
    
# b) A method total_value()
    def total_value(self):
       
        # returns the total market value of the portfolio
        total = 0
        for ticker, asset in self.holdings.items():
            total += asset["shares"] * asset["price_per_share"]
        return total
    
# c) A method most_valuable()
    def most_valuable(self):
       
        # returns the ticker of the asset with highest total value
        highest_ticker = None
        highest_value = 0
        for ticker, asset in self.holdings.items():
            asset_value = asset["shares"] * asset["price_per_share"]
            if asset_value > highest_value:
                highest_value = asset_value
                highest_ticker = ticker
        return highest_ticker
    
# d) A __str__ method
    def __str__(self):
        
        # returns a readable summary
        num_assets = len(self.holdings)
        total = self.total_value()
        return f"Portfolio Owner: {self.owner}, Assets: {num_assets}, Total Value: €{total:,.2f}"

# Create an instance and demonstrate all methods
portfolio = Portfolio("John Smith")

# Add at least three assets

# 10 shares of Apple at €150
portfolio.add_asset("AAPL", 10, 150)  
portfolio.add_asset("GOOGL", 5, 2800)  # 5 shares of Google at €2800
portfolio.add_asset("MSFT", 8, 300)  # 8 shares of Microsoft at €300

# Print output demonstration
print("=== Portfolio Demo ===")

print(portfolio)  
# uses __str__ method

print(f"Total portfolio value: €{portfolio.total_value():,.2f}")
print(f"Most valuable asset: {portfolio.most_valuable()}")
````

### Task 5: Importing a module (1p)

You are given the following list of order dates (as strings) and product names:

```python
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]
```

Using the `datetime` module, parse each date string into a `datetime.date` object. Compute and print the number of days between the earliest and latest order.

````python
# Importing a module 

from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

# Parse each date string into a datetime.date object
parsed_dates = []

for date_str, product in orders:
    
    # Convert string to datetime object
    date_obj = datetime.strptime(date_str, "%Y-%m-%d").date()
    parsed_dates.append(date_obj)
   
    #print
    print(f"Date: {date_obj}, Product: {product}")

# Find earliest and latest order
earliest = min(parsed_dates)
latest = max(parsed_dates)

# Compute days between
days_difference = (latest - earliest).days

# print
print(f"\nEarliest order: {earliest}")
print(f"Latest order: {latest}")
print(f"Days between earliest and latest order: {days_difference}")
````

### Task 6: Array Operations and Broadcasting (1p)

A company tracks quarterly revenue (in thousands of euros) across four quarters over three years:

```python
# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])
```

a) Compute and print the total annual revenue for each year (i.e., row sums). Use the `axis` parameter.

b) Compute the mean quarterly revenue across all years (i.e., column means).

c) Using broadcasting, apply a growth adjustment to the 2024 row by multiplying it element-wise with the array `[1.05, 1.03, 1.08, 1.02]`. Print the adjusted values.

d) Use boolean indexing to find and print all quarterly revenue values that exceed 250 (thousands of euros).

````python
# Array Operations and Broadcasting 

import numpy as np

# Revenue data
revenue = np.array([
    [210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])

# a) Compute total annual revenue for each year (row sums)
annual_revenue = revenue.sum(axis=1)

#print
print(f"Total annual revenue for each year: {annual_revenue}")
print(f"2022: €{annual_revenue[0]}k, 2023: €{annual_revenue[1]}k, 2024: €{annual_revenue[2]}k")

# b) Compute mean quarterly revenue across all years (column means)
quarterly_mean = revenue.mean(axis=0)

#print 
print(f"Mean quarterly revenue across all years: {quarterly_mean}")

# c) Apply growth adjustment to 2024 row using broadcasting
growth_adjustment = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth_adjustment

#print
print(f"2024 adjusted values: {adjusted_2024}")

# d) Boolean indexing - find all revenue values exceeding 250
exceeds_250 = revenue[revenue > 250]

#print
print(f"Quarterly revenue values exceeding €250k: {exceeds_250}")
````

### Task 7: Pandas (1p)

Create the following DataFrame:

```python
data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}
```

a) Add a new column `salary_per_year_exp` computed as `salary / years_exp`. Round to the nearest integer.

b) Sort the DataFrame by `performance_score` in descending order and print the top 3 employees.

c) Using `.loc`, select all employees in the Finance department with a performance score above 4.0. Print the result.

d) Rename the column `years_exp` to `experience_years` using the `.rename()` method.

e) Using `.iloc`, select and print the salary and performance score columns for the first four employees.

````python
# Pandas
import pandas as pd

data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]
}

# Create DataFrame
df = pd.DataFrame(data)

# a) Add new column salary_per_year_exp
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round(0).astype(int)

# print
print("DataFrame with salary_per_year_exp:")
print(df)

# b) Sort by performance_score descending, print top 3
top_3 = df.sort_values("performance_score", ascending=False).head(3)

# print
print("Top 3 employees by performance score:")
print(top_3)

# c) Using .loc, select Finance employees with performance > 4.0
finance_high_performers = df.loc[(df["department"] == "Finance") & (df["performance_score"] > 4.0)]

# print
print("Finance employees with performance score > 4.0:")
print(finance_high_performers)

# d) Rename years_exp to experience_years
df_renamed = df.rename(columns={"years_exp": "experience_years"})

# print
print("After renaming 'years_exp' to 'experience_years':")
print(df_renamed.columns.tolist())

# e) Using .iloc, select salary and performance_score for first 4 employees
first_four = df.iloc[0:4, [3, 4]]  # columns 3 and 4 are salary and performance_score

# print
print("Salary and performance score for first 4 employees:")
print(first_four)
````

### Task 8: More Pandas (1p)

A company records daily website traffic over four weeks. Generate the data by running the following script first:

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

Using this generated data, perform the following:

a) Create a DataFrame with columns: `date`, `visitors`, `conversion_rate`, and `avg_order_value`. Set `date` as the index.

b) Compute two new columns: `conversions` (= visitors × conversion_rate, rounded to the nearest integer) and `daily_revenue` (= conversions × avg_order_value).

````python
import numpy as np
import pandas as pd

# Generate data
rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

# a) Create DataFrame with columns: date, visitors, conversion_rate, avg_order_value

# Set date as index
df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

df.set_index("date", inplace=True)

# print
print("a) DataFrame with date as index:")

# Show first 5 rows
print(df.head(5))  

# b) Compute conversions and daily_revenue
df["conversions"] = (df["visitors"] * df["conversion_rate"]).round(0).astype(int)
df["daily_revenue"] = (df["conversions"] * df["avg_order_value"]).round(2)

# print
print("\nb) DataFrame with conversions and daily_revenue:")

# Show first 10 rows
print(df.head(10))  

print("\nSummary statistics:")
print(f"Total visitors: {df['visitors'].sum()}")
print(f"Total conversions: {df['conversions'].sum()}")
print(f"Total revenue: €{df['daily_revenue'].sum():,.2f}")
````
