````python
### Task 1: A list of dictionaries (1p)

You are given a list of raw transaction records, where each record is a dictionary:

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

Write a script that performs the following steps:

a) Filter out invalid transactions where the price is zero or the quantity is negative.

b) For each remaining transaction, compute a new field `total` = `price * quantity`.

c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

d) Print the region with the highest total revenue and its value using an f-string.
````

````python
# a) Filter out invalid transactions where the price is zero or the quantity is negative.

transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]

# create an empty list for valid transactions
valid_transactions = []

# go through each transaction
for transaction in transactions:
    
    # check if transaction is valid
    if transaction["price"] !=0 and transaction["quantity"] >= 0:
        
        # add valid transaction to the list
        valid_transactions.append(transaction)

print (valid_transactions)
````

````python
# b) for each remaining transaction, compute a new field `total` = `price * quantity`.

# loop through the valid transactions
for transaction in valid_transactions:

    # calculate 
    transaction["total"] = transaction["price"] * transaction["quantity"]

for transaction in valid_transactions:
    print(transaction)
````

````python
# c) Group the valid transactions by `region` and compute the total revenue per region. Store the result in a dictionary.

# create a new dictionary for revenue per region
revenue_per_region = {}

# loop through valid transactions and group them by region
for transaction in valid_transactions:

    # explanation: If this region has never been seen before
        #register it on the scoreboard with 0 first so Python doesn't crash with a KeyError 
    if transaction["region"] not in revenue_per_region:
        revenue_per_region[transaction["region"]] = 0
    
    # '+=' add to existing value --> revenue_per_region[transaction["region"]] = revenue_per_region[transaction["region"]] + transaction["total"]
    revenue_per_region[transaction["region"]] += transaction["total"]


# d) Print the region with the highest total revenue and its value using an f-string.

print (revenue_per_region)

# find the REGION with the highest revenue
highest_region = max(revenue_per_region, key=revenue_per_region.get)
    # max(revenue_per_region, : find the maximum through the dictionary revenue_per_region
    # key=revenue_per_region.get : compare by VALUE, not region NAME

# find the highest REVENUE
highest_revenue = revenue_per_region[highest_region]

# print the region with the highest total revenue and its value using an f-string
print(f"The highest revenue region is {highest_region} with {highest_revenue}")
    # print(f"...") : print value using an f-string
````

### Task 2: Creating a function (1p)

You are given the following list of daily sales figures (in euros) for a two-week period:

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

Write a function called `summarise_sales` that takes a list of daily sales values and returns a dictionary containing:

`"total"`: the sum of all daily sales

`"average"`: the mean daily sales value, rounded to two decimals

`"best_day"`: the index (starting from 1) of the day with the highest sales

`"worst_day"`: the index (starting from 1) of the day with the lowest sales

`"above_average_count"`: the number of days where sales exceeded the average

Call the function with the list above and print the returned dictionary.

````python
# write a function

def summarise_sales(sales):
    """list of daily sales"""
    
    # calculate total daily sales, store in variable 'total'
    total = sum(sales)

    #calculate average daily sales, total divided by number of days, round the answer to '2' decimal places
    average = round(sum(sales) / len(sales), 2)

    # find the best day with the highest sales
        # .index() = find which position that number is at 
        # max(sales) = find the biggest number in the list
        # add 1 because days start from 1, not 0
    best_day = sales.index(max(sales)) + 1

    # find the worst day which has the lowest sales
    worst_day = sales.index(min(sales))+1

    # find above average: how many days had sales higher than the average
    above_average_count = 0
    for daily_amount in sales:
        if daily_amount > average:
            # for every daily sales that is above average, add 1 to the counter
            above_average_count += 1

# Return all results as a dictionary

    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }
    
````

````python
# Call the function and print the returned dictionary

# Call the function with the sales data
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

result = summarise_sales(sales)

# Print the entire returned dictionary
print(result)
````
